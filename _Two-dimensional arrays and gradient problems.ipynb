```python
import numpy as np
import matplotlib.pyplot as plt

# Create an ndarray for x with values at 0.1 intervals from -50 to 50
x_values = np.arange(-50, 50.1, 0.1)

# Calculate y for each element of x using the formula y = 0.5 * x + 1
y_values = 0.5 * x_values + 1

# Calculate the gradients
gradients = np.diff(y_values) / np.diff(x_values)

# Plotting the linear function y = 0.5 * x + 1
plt.figure(figsize=(10, 5))

# Plot 1: Linear function
plt.subplot(1, 2, 1)
plt.plot(x_values, y_values, label='y = 0.5 * x + 1')
plt.title('Graph of y = 0.5 * x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Plot 2: x vs gradient
plt.subplot(1, 2, 2)
plt.plot(x_values[:-1], gradients, label='Gradient')
plt.title('Gradient vs x')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.legend()

# Adjust layout for better visualization
plt.tight_layout()

# Display the plots
plt.show()
```

# PROBLEM 5

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_gradients_and_plot(equation, start, end, interval):
    # Create an ndarray for x with values at specified intervals
    x_values = np.arange(start, end + interval, interval)

    # Calculate y for the given equation
    if equation == "x^2":
        y_values = x_values ** 2
    elif equation == "2x^2 + 2x":
        y_values = 2 * x_values ** 2 + 2 * x_values
    elif equation == "sin(x^(1/2))":
        y_values = np.sin(np.sqrt(x_values))

    # Calculate the gradients
    gradients = np.diff(y_values) / np.diff(x_values)

    # Plotting the graphs
    plt.figure(figsize=(12, 5))

    # Plot 1: Graph of the equation
    plt.subplot(1, 2, 1)
    plt.plot(x_values, y_values, label=f'y = {equation}')
    plt.title(f'Graph of y = {equation}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()

    # Plot 2: x vs gradient
    plt.subplot(1, 2, 2)
    plt.plot(x_values[:-1], gradients, label='Gradient')
    plt.title('Gradient vs x')
    plt.xlabel('x')
    plt.ylabel('Gradient')
    plt.legend()

    # Adjust layout for better visualization
    plt.tight_layout()

    # Display the plots
    plt.show()

    return gradients

# Calculate gradients and plot the graphs for each equation
equation1_gradients = calculate_gradients_and_plot("x^2", -50, 50, 0.1)
equation2_gradients = calculate_gradients_and_plot("2x^2 + 2x", -50, 50, 0.1)
equation3_gradients = calculate_gradients_and_plot("sin(x^(1/2))", 0, 50, 0.1)

# Output the gradients for each equation
print("Gradients for y = x^2:")
print(equation1_gradients)

print("\nGradients for y = 2x^2 + 2x:")
print(equation2_gradients)

print("\nGradients for y = sin(x^(1/2)):")
print(equation3_gradients)
```

# PROBLEM 6

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_gradients_and_plot(equation, start, end, interval):
    # Create an ndarray for x with values at specified intervals
    x_values = np.arange(start, end + interval, interval)

    # Calculate y for the given equation
    if equation == "x^2":
        y_values = x_values ** 2
    elif equation == "2x^2 + 2x":
        y_values = 2 * x_values ** 2 + 2 * x_values
    elif equation == "sin(x^(1/2))":
        y_values = np.sin(np.sqrt(x_values))

    # Calculate the gradients
    gradients = np.diff(y_values) / np.diff(x_values)

    # Plotting the graphs
    plt.figure(figsize=(12, 5))

    # Plot 1: Graph of the equation
    plt.subplot(1, 2, 1)
    plt.plot(x_values, y_values, label=f'y = {equation}')
    plt.title(f'Graph of y = {equation}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()

    # Plot 2: x vs gradient
    plt.subplot(1, 2, 2)
    plt.plot(x_values[:-1], gradients, label='Gradient')
    plt.title('Gradient vs x')
    plt.xlabel('x')
    plt.ylabel('Gradient')
    plt.legend()

    # Find the minimum value of y and its index
    min_y_value = np.min(y_values)
    min_y_index = np.argmin(y_values)

    # Display the minimum y value and its corresponding x value
    print(f"Minimum y value: {min_y_value} at x = {x_values[min_y_index]}")

    # Calculate slope before and after minimum y
    slope_before_min = (y_values[min_y_index] - y_values[min_y_index - 1]) / (x_values[min_y_index] - x_values[min_y_index - 1])
    slope_after_min = (y_values[min_y_index + 1] - y_values[min_y_index]) / (x_values[min_y_index + 1] - x_values[min_y_index])

    print(f"Slope before minimum y: {slope_before_min}")
    print(f"Slope after minimum y: {slope_after_min}")

    # Adjust layout for better visualization
    plt.tight_layout()

    # Display the plots
    plt.show()

    return min_y_value

# Calculate minimum y value and plot the graphs for each equation
min_y_x2 = calculate_gradients_and_plot("x^2", -50, 50, 0.1)
min_y_2x2_2x = calculate_gradients_and_plot("2x^2 + 2x", -50, 50, 0.1)
min_y_sin = calculate_gradients_and_plot("sin(x^(1/2))", 0, 50, 0.1)

print("\nMinimum y value for y = x^2:", min_y_x2)
print("Minimum y value for y = 2x^2 + 2x:", min_y_2x2_2x)
print("Minimum y value for y = sin(x^(1/2)):", min_y_sin)
```
